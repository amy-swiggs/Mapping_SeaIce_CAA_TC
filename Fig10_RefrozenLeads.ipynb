```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ls8_data = pd.read_excel('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/GIS_Processing/LS8_RF_Leads_Stats/LS8_RF_Leads_Stats_Python.xlsx')
```

```python
fig, ax = plt.subplots(figsize=[7,7])
plt.rcParams.update({'font.size': 14})

# February
ax.errorbar(0.8,ls8_data.iloc[0,3],yerr=5.7,
                     color = 'darkcyan',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax.errorbar(1.2,ls8_data.iloc[0,5],yerr=5.7,
                     color = 'indianred',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')

# March
ax.errorbar(2.8,ls8_data.iloc[1,3],yerr=5.7,
                     color = 'darkcyan',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax.errorbar(3.2,ls8_data.iloc[1,5],yerr=5.7,
                     color = 'indianred',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')


# May
ax.errorbar(4.8,ls8_data.iloc[2,3],yerr=5.7,
                     color = 'darkcyan',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax.errorbar(5.2,ls8_data.iloc[2,5],yerr=5.7,
                     color = 'indianred',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')


# October
ax.errorbar(6.8,ls8_data.iloc[3,3],yerr=5.7,
                     color = 'darkcyan',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax.errorbar(7.2,ls8_data.iloc[3,5],yerr=5.7,
                     color = 'indianred',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')


ax.set_xticks([1,3,5,7])
ax.set_xticklabels(('February','March','May','October'))
ax.xaxis.grid(True,'major',alpha=0.2)
ax.yaxis.grid(True,'major',alpha=0.2)
ax.set_ylabel('Sea Ice Concentration (%)', fontsize = 16)
legenditems = ('Refrozen Lead Classification','Strict Lead Classification')
ax.legend(legenditems,fontsize=16,loc='lower left')

fig.savefig('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC_Plots/ReSubmission_Plots/Fig10_LS8_RF_Leads.png',dpi=300,bbox_inches="tight")
```
