In this code we map the annual, seasonal, and regional averages in sea ice concentration for CS2, NASA Team, and the Canadian Ice Service.

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.ticker as ticker
import glob
import pandas as pd
import os
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

# Monthly Means (Fig4a)

```python
# Read in dataframes containing the monthly means for the CAA

monthly_means = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/AllSensors_Monthly_Averages.csv')
```

```python
fig_a, ax_a = plt.subplots(figsize=[7,7])
plt.rcParams.update({'font.size': 14})

legend_handles, legend_labels = [],[]

for i, team_mean in enumerate(monthly_means['Team Means']):
    x_position = 1 + i # Introduce a counter so the bars can be plotted cleary along the x axis

    team_plot = ax_a.errorbar(x=x_position,y=team_mean,yerr = 5.5,
                     color = 'PaleVioletRed',capsize = 7,capthick = 1.8,
                        linewidth = 1.8,linestyle = 'none')

    if i == 0:
        legend_handles.append(team_plot)
        legend_labels.append('NASA Team')


for i, cis_mean in enumerate(monthly_means['CIS Means']):
    x_position = 0.8 + i
    
    cis_plot = ax_a.errorbar(x=x_position,y=cis_mean,yerr = 5,
                     color = 'DarkOrange',capsize = 7,capthick = 1.8,
                        linewidth = 1.8,linestyle = 'none')

    if i == 0:
        legend_handles.append(cis_plot)
        legend_labels.append('Canadian Ice Service')


for i, cs2_mean in enumerate(monthly_means['CS2 Means']):
    x_position = 1.2 + i
    
    cs2_plot = ax_a.errorbar(x=x_position,y=cs2_mean,yerr = 5,
                     color = 'SteelBlue',capsize = 7,capthick = 1.8,
                        linewidth = 1.8,linestyle = 'none')
    
    if i == 0:
        legend_handles.append(cs2_plot)
        legend_labels.append('CryoSat-2')


ax_a.set_xlabel('Month', fontsize = 18)
ax_a.set_ylim(30, 105)
ax_a.xaxis.grid(True,'major',alpha=0.2)
ax_a.yaxis.grid(True,'major',alpha=0.2)
ax_a.set_xticks([1,2,3,4,5,6,7,8,])
ax_a.set_xticklabels(['Oct','Nov','Dec','Jan','Feb','Mar','Apr','May'])
ax_a.set_ylabel('Sea Ice Concentration (%)', fontsize = 18)

plt.yticks(fontsize = 16)
plt.xticks(fontsize = 16)

ax_a.legend(legend_handles, legend_labels, fontsize=16, loc='lower right')

plt.tight_layout()
plt.show()
```

# Annual Means

```python
# Read annual means
team_annual_df = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/NASA_Team_Annual_Means.csv')

cs2_annual_df = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CS2_Annual_Means.csv')

cis_annual_df = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CIS_Annual_Means.csv')
```

```python
fig_b, ax_b = plt.subplots(figsize=[14,7])
plt.rcParams.update({'font.size': 14})

team_plot, = ax_b.plot(team_annual_df['Years'], team_annual_df['Mean'],linewidth = 2.5,color='PaleVioletRed')
ax_b.fill_between(team_annual_df['Years'], team_annual_df['Mean'] - 5,
                team_annual_df['Mean'] + 5, color='PaleVioletRed', alpha=0.1)

cis_plot, = ax_b.plot(cis_annual_df['Years'], cis_annual_df['Mean'],linewidth = 2.5,color='DarkOrange')
ax_b.fill_between(cis_annual_df['Years'], cis_annual_df['Mean'] - 5,
                cis_annual_df['Mean'] + 5, color='DarkOrange', alpha=0.1)

cs2_plot, = ax_b.plot(cs2_annual_df['Years'], cs2_annual_df['Mean'],linewidth = 2.5,color='SteelBlue')
ax_b.fill_between(cs2_annual_df['Years'], cs2_annual_df['Mean'] - 5.5,
                cs2_annual_df['Mean'] + 5.5, color='SteelBlue', alpha=0.1)

ax_b.set_xlabel('Year', fontsize = 18)
ax_b.set_ylabel('Sea Ice Concentration (%)', fontsize = 18)
ax_b.set_ylim(60, 97)
ax_b.xaxis.grid(True,'major',alpha=0.2)
ax_b.yaxis.grid(True,'major',alpha=0.2)

ax_b.legend([team_plot, cis_plot, cs2_plot], 
            ['NASA Team', 'Canadian Ice Service', 'CryoSat-2'],
            fontsize=16, loc='lower right')

plt.yticks(fontsize = 16)
plt.xticks(fontsize = 16)

plt.show()
```

```python
fig_c, ax_c = plt.subplots(figsize=[7,7])
plt.rcParams.update({'font.size': 14})

team_x = [0.9,1.9,2.9]
team_plot = ax_c.errorbar(team_x,team_regional_means_df['Mean'],5,capsize=7,color='PaleVioletRed', fmt=' ',
                       capthick = 1.8,linewidth = 1.8,linestyle = 'none')

cis_x = [0.7,1.7,2.7]
cis_plot = ax_c.errorbar(cis_x,cis_regional_means_df['Mean'],5.5,capsize=7,color='DarkOrange', fmt=' ',
                       capthick = 1.8,linewidth = 1.8,linestyle = 'none')

cs2_x = [1.1,2.1,3.1]
cs2_plot = ax_c.errorbar(cs2_x,cs2_regional_means_df['Mean'],5.5,capsize=7,color='SteelBlue', fmt=' ',
                       capthick = 1.8,linewidth = 1.8,linestyle = 'none')


ax_c.set_xticks([1,2,3])
ax_c.set_xlim(0.5,3.5)
ax_c.xaxis.set_major_formatter(ticker.FixedFormatter(['Beaufort Sea', 'Central CAA','Baffin Bay']))
ax_c.xaxis.grid(True,'major',alpha=0.2)
ax_c.yaxis.grid(True,'major',alpha=0.2)
ax_c.set_ylabel('Sea Ice Concentration (%)', fontsize = 18)
ax_c.set_xlabel('Region', fontsize = 18)

plt.yticks(fontsize = 16)
plt.xticks(fontsize = 16)

plt.tight_layout()
plt.legend(labels=('NASA Team','Canadian Ice Service','CryoSat-2'),fontsize=16,loc='lower left')
plt.show()
```
