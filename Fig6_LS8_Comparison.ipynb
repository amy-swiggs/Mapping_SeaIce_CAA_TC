This code produces four plots, comparing the sea ice concentration values at LS8 intersecting locations, and for the full record from CS2, CIS, and NASA Team

```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.ticker as ticker
import glob
import pandas as pd
```

```python
# Intersecting locations
# 
# Monthly averages at intersecting locations
intersecting_averages = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/LS8_Comparison_Intersection_Locations_MonthlyAverages.csv')

# All values at intersecting locations
ls8_intersections = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/LS8_Comparison_Intersection_Locations.csv')
team_intersections = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/NASA_Team_Comparison_Intersection_Locations.csv')
cis_intersections = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CIS_Comparison_Intersection_Locations.csv')
cs2_intersections = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CS2_Comparison_Intersection_Locations.csv')

# Full Record
full_record_monthly_averages = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/LS8_Comparison_Full_Record_MonthlyAverages.csv')

# All values for the full record
team_means_all = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/NASA_Team_All_Monthly_Averages.csv')
cs2_means_all = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CS2_All_Monthly_Averages.csv')
cis_means_all = pd.read_csv('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC/Mapping_SIC_TC/CIS_All_Monthly_Averages.csv')
```

```python
fig, ax = plt.subplots(2, 2, figsize=[20,16])
plt.rcParams.update({'font.size': 14})

# Top Left Plot

ax[0,0].errorbar([1.2,3.2,5.2,7.2,9.2],intersecting_averages['Team Means'],yerr=5,
                     color = 'PaleVioletRed',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax[0,0].errorbar([1.4,3.4,5.4,7.4,9.4],intersecting_averages['CIS Means'],yerr=5.5,
                     color = 'DarkOrange',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax[0,0].errorbar([1.6,3.6,5.6,7.6,9.6],intersecting_averages['CS2 Means'],yerr=5,
                     color = 'SteelBlue',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax[0,0].errorbar([1.8,3.8,5.8,7.8,9.8],intersecting_averages['LS8 Means'],yerr=5.7,
                     color = 'MediumSeaGreen',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')

ax[0,0].set_xticks([1,3,5,7,9])
ax[0,0].set_xticklabels(('February','March','April','May','October'))
ax[0,0].set_ylim(30, 105)
ax[0,0].xaxis.grid(True,'major',alpha=0.2)
ax[0,0].yaxis.grid(True,'major',alpha=0.2)
ax[0,0].set_ylabel('Sea Ice Concentration (%)', fontsize = 16)
ax[0,0].set_title('LS8 Intersection Locations',fontsize=18)
legenditems = ('NASA Team','Canadian Ice Service','CryoSat-2','Landsat 8')
ax[0,0].legend(legenditems,fontsize=16,loc='lower left')

# Bottom Left Plot
bins5 = range(50,105,5)
xlim = ([50,100])
ylim = ([0,105])

ax[1,0].hist(team_intersections['Mean'],histtype = 'step', bins=bins5,
             edgecolor='PaleVioletRed',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,0].hist(cis_intersections['Mean'],histtype = 'step', bins=bins5,
             edgecolor='DarkOrange',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,0].hist(cs2_intersections['Mean'],histtype = 'step', bins=bins5,
             edgecolor='SteelBlue',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,0].hist(ls8_intersections['Mean'],histtype = 'step', bins=bins5,
             edgecolor='MediumSeaGreen',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)

ax[1,0].set_xlim(xlim)
ax[1,0].set_ylim(ylim)
ax[1,0].xaxis.grid(True,'major',alpha=0.2)
ax[1,0].yaxis.grid(True,'major',alpha=0.2)
ax[1,0].set_ylabel('Count', fontsize = 16)
ax[1,0].set_title('LS8 Intersection Locations',fontsize=18)
ax[1,0].set_xlabel('Sea Ice Concentration (%)', fontsize = 16)
ax[1,0].legend(labels = ['NASA Team','Canadian Ice Service','CryoSat-2','Landsat 8'],fontsize=16,loc='upper left')

# Top Right Plot

ax[0,1].errorbar([0.8,2.8,4.8,6.8,8.8],full_record_monthly_averages['Team Means'],yerr=5,
                     color = 'PaleVioletRed',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax[0,1].errorbar([1.2,3.2,5.2,7.2,9.2],full_record_monthly_averages['CIS Means'],yerr=5,
                     color = 'DarkOrange',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')
ax[0,1].errorbar([1,3,5,7,9],full_record_monthly_averages['CS2 Means'],yerr=5.5,
                     color = 'SteelBlue',capsize = 5,capthick = 1.8,linewidth = 1.8,linestyle = 'none')

ax[0,1].set_xticks([1,3,5,7,9])
ax[0,1].set_xticklabels(('February','March','April','May','October'),fontsize=14)
ax[0,1].set_ylim(30, 105)
ax[0,1].xaxis.grid(True,'major',alpha=0.2)
ax[0,1].yaxis.grid(True,'major',alpha=0.2)
ax[0,1].set_ylabel('Sea Ice Concentration (%)', fontsize = 16)
ax[0,1].set_title('Total CAA Record',fontsize=18)
legenditems = ('NASA Team','Canadian Ice Service','CryoSat-2')
ax[0,1].legend(legenditems,fontsize=16,loc='lower left')

# Bottom Right Plot

ax[1,1].hist(team_means_all['Mean'],histtype = 'step', bins=bins5,
             edgecolor='PaleVioletRed',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,1].hist(cis_means_all['Mean'],histtype = 'step', bins=bins5,
             edgecolor='DarkOrange',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,1].hist(cs2_means_all['Mean'],histtype = 'step', bins=bins5,
             edgecolor='SteelBlue',facecolor='None',linewidth=2,
             alpha=0.9, cumulative=True)
ax[1,1].xaxis.grid(True,'major',alpha=0.2)
ax[1,1].yaxis.grid(True,'major',alpha=0.2)
ax[1,1].set_xlim(xlim)
ax[1,1].set_ylim(ylim)
ax[1,1].set_ylabel('Count', fontsize = 16)
ax[1,1].set_xlabel('Sea Ice Concentration (%)', fontsize = 16)
ax[1,1].set_title('Total CAA Record',fontsize=18)
ax[1,1].legend(labels = ['NASA Team','Canadian Ice Service','CryoSat-2'],fontsize=16,loc='upper left')


plt.text(-9.5,233,'(a)',fontsize=14)
plt.text(50.7,233,'(b)',fontsize=14)
plt.text(-9.5,107,'(c)',fontsize=14)
plt.text(50.7,107,'(d)',fontsize=14)

plt.show()

fig.savefig('/Users/amyswiggs/Desktop/Autumn2023/Chapter2/Northumbria_Spring_Data/NWP_SIC_Plots/Submission_Plots/Fig06.png',dpi=400,bbox_inches="tight")


```
