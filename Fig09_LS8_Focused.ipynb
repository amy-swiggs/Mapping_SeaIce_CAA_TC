In this code, four Landsat 8 sea ice concentration images are plotted with near-coincident CS2, NT, and CIS data for a focused evaluation in Figure 9.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import ListedColormap
import matplotlib.colors
from matplotlib.colors import LinearSegmentedColormap
import rasterio
from rasterio.transform import Affine
from rasterio.warp import calculate_default_transform, reproject, Resampling
from rasterio.plot import show
import cartopy
import cartopy.crs as ccrs
import geopandas as gpd
from shapely.prepared import prep
from shapely.geometry import Point
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib.font_manager import FontProperties
```

```python
# Define the custom Lambert Azimuthal Equal Area projection
map_projection = ccrs.LambertAzimuthalEqualArea(central_longitude=-105,central_latitude=74)

# Coordinates to transform (lon, lat) in the Plate Carree projection
geo_extent = [-140, -70, 64, 79]

def transform_coords(lon, lat, projection):
    x, y = projection.transform_point(lon, lat, ccrs.PlateCarree())
    return x, y

# Get corners of 'bounding box' from geo_extent
corners = [
    (geo_extent[0], geo_extent[2]),  # SW corner
    (geo_extent[1], geo_extent[2]),  # SE corner
    (geo_extent[0], geo_extent[3]),  # NW corner
    (geo_extent[1], geo_extent[3])   # NE corner
]

# Transform coordinates to the desired map area
proj_corners = [transform_coords(lon, lat, map_projection) for lon, lat in corners]
x_coords, y_coords = zip(*proj_corners)

x_min, x_max = min(x_coords), max(x_coords)
y_min, y_max = min(y_coords), max(y_coords)

grid_spacing = 25000 # 25 km

x_grid_edge = np.arange(x_min-grid_spacing,x_max + grid_spacing, grid_spacing)
y_grid_edge = np.arange(y_min-grid_spacing,y_max + grid_spacing, grid_spacing)
new_x_edge, new_y_edge= np.meshgrid(x_grid_edge, y_grid_edge) # For CS2 data

print('Edges grid shape = ',np.shape(new_x_edge)) # edges should be one greater than centres

x_grid_centre = np.arange(x_min-grid_spacing/2,x_max + grid_spacing/2, grid_spacing)
y_grid_centre = np.arange(y_min-grid_spacing/2,y_max + grid_spacing/2, grid_spacing)
new_x_centre, new_y_centre = np.meshgrid(x_grid_centre, y_grid_centre) # For PMV data

# Set up the figure
fig = plt.figure(figsize=(15, 20))
ax = plt.axes(projection=map_projection)
ax.coastlines(linewidth=0.2)
ax.scatter(new_x_edge,new_y_edge,0.05, color='blue')
plt.show()
```

```python
def polygon_masking(grid_x,grid_y,shapefile):
    x_flt = grid_x.flatten()
    y_flt = grid_y.flatten()

    # Prepare shapefiles for processing
    geom = shapefile.geometry.union_all()
    p_geom = prep(geom)

    # Use flattened x and y coordinates to find points within the polygon
    mask_flt = np.array([p_geom.contains(Point(x, y)) for x, y in zip(x_flt, y_flt)])
    # Reshape the mask back to the original array shape
    mask = mask_flt.reshape(grid_x.shape)

    return(mask)


def add_shapefile_to_plot(ax, gdf,color,edgecolor, linewidth):
    for geom in gdf.geometry:
        ax.add_geometries([geom], crs=map_projection, facecolor=color,edgecolor=edgecolor, linewidth=linewidth)

def quick_CIS_processing(cis_gdf):

    # Reformat NSIDC CIS values
    cis_gdf['CT_numeric'] = pd.to_numeric(cis_gdf['CT'], errors='coerce')
    cis_gdf.loc[cis_gdf['CT_numeric'] == 92, 'CT_numeric'] = 100
    cis_gdf.loc[cis_gdf['CT_numeric'] == 91, 'CT_numeric'] = 95
    cis_gdf.loc[cis_gdf['CT_numeric'] == 2, 'CT_numeric'] = 0
    cis_gdf.loc[cis_gdf['CT_numeric'] == 1, 'CT_numeric'] = 0
    cis_gdf.loc[cis_gdf['CT_numeric'] == 55, 'CT_numeric'] = 0 

    return(cis_gdf)
```

```python
# Create colorbar for NT and CS2

cvals  = [0,1,1.7,2.4,2.6,2.7,2.75]
colors = ['#07146e','#3644a6','#5d6ac2','#949ddb','#c3c7e8','#eceefa','#f7f8ff']

norm=plt.Normalize(min(cvals),max(cvals))
tuples = list(zip(map(norm,cvals), colors))
new_cmap = matplotlib.colors.LinearSegmentedColormap.from_list("", tuples)

# Colorbar for LS8 reversed if required
cvals  = [0,1,1.7,2.4,2.6,2.7,2.75]
colors = ['#f7f8ff','#eceefa','#c3c7e8','#949ddb','#5d6ac2','#3644a6','#07146e']

norm=plt.Normalize(min(cvals),max(cvals))
tuples = list(zip(map(norm,cvals), colors))
new_cmap_r = matplotlib.colors.LinearSegmentedColormap.from_list("", tuples)
```

```python
def plotting_sic_map(shapefile,x,y,data,vmin,vmax,extentminx,extentmaxx,extentminy,extentmaxy):
    fig = plt.figure(figsize=(10, 10))
    ax = plt.axes(projection=map_projection)
    ax.coastlines(linewidth=0.2,zorder=1.5)
    ax.add_feature(cartopy.feature.LAND, color=('#396e1d'), zorder=2)
    ax.add_feature(cartopy.feature.OCEAN, color=('darkgrey'), edgecolor='black',alpha=0.7,zorder=0)
    for geom in shapefile.geometry:
        ax.add_geometries([geom], crs=map_projection, facecolor='none',edgecolor='#d7354d', linewidth=1,alpha=1,zorder=3)
    
    pcm = ax.pcolormesh(x,y,data,cmap=new_cmap,transform=map_projection,vmin=vmin,vmax=vmax,zorder=1)
    ax.set_extent([extentminx, extentmaxx, extentminy, extentmaxy], crs=map_projection)

    return(fig)
```
